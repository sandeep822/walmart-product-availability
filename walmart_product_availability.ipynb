```python
# Extract domain from 'Product Url'
df['Domain'] = df['Product Url'].apply(lambda x: x.split('/')[2])
df.head()
```

```python
# 5. Dropping Unnecessary Columns
df.drop(['Uniq Id', 'Gtin'], axis=1, inplace=True)



df.head()
```

```python


# 6. Duplicate Removal
df.drop_duplicates(inplace=True)
df.head()
```

```python
# 7. Boolean Values (Convert Text to Boolean)
df['Available'] = df['Available'].astype(bool)
df.head()
```

```python
# 8. Standardize Brand Names (Title Case)
df['Brand'] = df['Brand'].str.title()

df.head()
```

```python
# 9. Text Length Validation (Filter Short Descriptions)
df = df[df['Description'].str.len() > 10]
df.head()
```

```python

# 10. Label Encoding (Encode Categorical Variables)
label_encoder = LabelEncoder()
df['Brand'] = label_encoder.fit_transform(df['Brand'])
df.head()
```

```python

# 11. Standard Scaling (Scale Numeric Features)
scaler = StandardScaler()
numeric_features = ['List Price', 'Sale Price']
df[numeric_features] = scaler.fit_transform(df[numeric_features])
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
from sklearn.impute import SimpleImputer

# Assuming X_train, X_test, y_train, y_test are defined

# Imputation
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Create and train different classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'SVM': SVC(probability=True),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Neural Network': MLPClassifier(max_iter=1000)
}

results = {'Model': [], 'Accuracy': [], 'ROC/AUC': []}

# Plot ROC curves
plt.figure(figsize=(10, 6))

for model_name, model in classifiers.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test_imputed)
    accuracy = accuracy_score(y_test, y_pred)

    results['Model'].append(model_name)
    results['Accuracy'].append(accuracy)

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test_imputed)[:, 1]
        roc_auc = roc_auc_score(y_test, y_prob)
        results['ROC/AUC'].append(roc_auc)

        # Compute ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_prob)

        # Plot ROC curve
        plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

    else:
        results['ROC/AUC'].append(None)  # Some classifiers may not have predict_proba method

plt.plot([0, 1], [0, 1], linestyle='--', color='grey', label='Random')
plt.title('ROC Curves for Classifiers')
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# Create a DataFrame from the results dictionary
results_df = pd.DataFrame(results)

# Print the results DataFrame
print(results_df)
```
